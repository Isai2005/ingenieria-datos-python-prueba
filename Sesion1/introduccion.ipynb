# Listas

```python
# listas

lista1 = []                                   # lista vacia

lista2 = [3, 5, 8]                            # lista con valores int (enteros)

lista3 = ['Manzana', 'Platano']               # lista con valores str (string)

lista4 = [3, 'Manzana', 3.4]                  # lista con varios tipos de datos

lista5 = [3, 'Manzana', [1, 5, 'Platano']]    # lista anidada (podemos tener listas dentro de otra lista)

print(lista5)
```

```python
# operador indice []

print(lista5[2])                            # para ingresar a un elemento en especifico 
```

```python
# para obtener elementos dentro la lista anidada

print(lista5[2][2])
```

```python
# para obtener el ultimo elemento de la lista para cuando son muchos

print(lista5[-1])
```

```python
lista6 = [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
print(lista6[2:5])       # para acceder a la lista a un valor especifico,usamos rangos de aonde a onde queremos ir  y los puntos : 
```

```python
# obtener un indice y que ponga los que siguen seguimos usando el rango y los dos puntos :
print(lista6[5:])
```

```python
# como modificar elementos en listas

lista6[4] = 10
print(lista6)
```

```python
# agregar nuevos elementos a las listas, usamos el metodo append

lista6.append(11)
print(lista6)

```

```python
# creamos una lista para prueba

lista_aux = [12, 13, 14]
```

```python
# si agregamos esa lista con el metodo append para que se agregue a la lista6, se agregara como si fuera lista anidada, algo que no queremos

lista6.append(lista_aux)
print(lista6)
```

```python
# para que se agrege secuencialmente y no como anidada usamos la palabra especial 'extend'

lista6.extend(lista_aux)
print(lista6)
```

```python
# como eliminar elementos en una lista

lista6.remove([12, 13, 14])
print(lista6)
```

# Tuplas

```python
# son muy parecidas a las listas pero las listas van con corchetes y las tuplas con parentesis
# tuplas de una vez que las creas ya no las puedes modificar

tupla = ()

tupla1 = (3, 4, 5)

tupla2 = ('Manzana', 'Platano')

tupla3 = (1, 3, 'Manzana')

tupla4 = ('Manzana', 2, 4, (3, 4, 5))

```

```python
# para acceder a las tuplas usamos lo mismo a las listas, los indices

print(tupla4[0])
```

# Diccionarios

```python
# se crean entre llaves y los elementos van separado por coma [los diccionarios conllevan clave y valor]

diccionario = {}

diccionario1 = {
    1:'Manzana', 2:'Balon'
}

diccionario2 = {
    'nombre':'isai', 1: [2, 3, 4]
}

diccionario3 = {
    'nombre':'juan', 'edad':20, 'telefono': 9982981654
}



```

```python
# para hacer operaciones ya no usamos los indices, usamos las claves, los keys

print(diccionario3['nombre'])
```

```python
# podemos agregar valores a los diccionarios

diccionario3['direccion'] = 'calle ni un brillo'
```
