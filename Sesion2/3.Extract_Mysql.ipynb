```python
import sqlalchemy as db
import pandas as pd

```

```python
# Como conectarnos a MySQL

engine = db.create_engine('mysql+mysqlconnector://root:root@localhost:3310/retail_db')

conn = engine.connect()
```

```python
# Para leer tablas completas  de MySQL usamos el read_sql_table

df_departments = pd.read_sql_table('departments', con=conn)
df_departments
```

```python
# para leer algo en especifico como por ejemplo consultas usamos el read_sql_query 

df_departments = pd.read_sql_query("SELECT * FROM departments", con=conn)
df_departments
```
